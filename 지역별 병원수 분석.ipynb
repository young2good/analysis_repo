```python
temp2 = temp[temp['표시과목'] == '계']
```

```python
temp_val2 = pd.merge(temp, temp2, how='left',on='시군구명칭')
```

```python
temp_val2['ratio'] = temp_val2['개수_x'] / temp_val2['개수_y']
temp_val2.head()
```

```python
new_temp_val = temp_val2[temp_val2['표시과목_x'] !='계']
new_temp_val[['시군구명칭','표시과목_x','개수_x','개수_y','ratio']]
```

```python
new_temp_val = new_temp_val[['시군구명칭','표시과목_x','개수_x','개수_y','ratio']]
new_temp_val.columns = ['시군구명칭','표시과목','개수','전체개수','ratio']
```

```python
new_temp_val.head()
```

```python
(
    alt.
    Chart(new_temp_val).
    encode(
        x=alt.X('시군구명칭'),
        y=alt.Y('표시과목').sort('-color'),
        color='ratio'
    ).
    mark_rect()
)
```

```python
new_temp_val.head()
```

```python
(
    alt.
    Chart(new_temp_val).
    transform_window(
        rank = 'rank()',
        sort = [
            alt.SortField('ratio', order = 'descending')
            ],
        groupby = ['시군구명칭']
    ).
    transform_filter(
        alt.datum.rank <= 5
    ).
        encode(
        x=alt.X('ratio:Q'),
        y=alt.Y('시군구명칭'),
        color = '표시과목',
        tooltip = '표시과목',
        # y = alt.Y('시군구명칭').sort(order = 'ascending')
        order = alt.Order(
            'ratio',
            sort = 'descending'
        )
    ).
    mark_bar()
)
```

```python
## 기간 데이터 가져옴

df_temp = pd.read_excel('E:\download\시군구별_표시과목별_의원_현황_20240826070312.xlsx')
```

```python
df_filled = df_temp.fillna(method='ffill')
```

```python
df_filled.head()
```

```python
df_filled.columns = ['시도명', '시군구명','표시과목','시점','병원수']
```

```python
df_filled[df_filled['시도명'] == '서울']['시군구명'].unique()
```

```python
df_new_temp = df_filled[(df_filled['시도명'] == '서울') &(df_filled['시군구명'] == '소계') & (df_filled['표시과목'] != '계')]
```

```python
(
    alt.
    Chart(df_new_temp).
    transform_filter(
        alt.FieldOneOfPredicate(field = '표시과목', oneOf= ['소아청소년과', '외과'])
    ).
    encode(
        x = '시점',
        y = '병원수',
        color = '표시과목'
    ).
    mark_line()
)
```
