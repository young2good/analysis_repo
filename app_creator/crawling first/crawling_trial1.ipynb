```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup
import time

url = "https://cafe.daum.net/_c21_/bbs_read?grpid=1O7ju&fldid=IxUe&datanum=234615"

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install())
)

# 게시글 접속
driver.get(url)

# 직접 로그인
input("로그인 완료 후 엔터 누르세요")

# 페이지 다시 로드
driver.get(url)

# 렌더링 대기
time.sleep(3)

html = driver.page_source

soup = BeautifulSoup(html, "html.parser")

# =========================
# 제목 추출
# =========================

title = soup.find("meta", {"property": "og:title"})

if title:
    print("제목:")
    print(title["content"])

# =========================
# 본문 요약 추출
# =========================

description = soup.find("meta", {"property": "og:description"})

if description:
    print("\n본문:")
    print(description["content"])

driver.quit()
```

```python
### 마지막 트라이얼
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup
import time

url = "https://cafe.daum.net/_c21_/bbs_read?grpid=1O7ju&fldid=IxUe&datanum=234613"

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install())
)

driver.get(url)

input("로그인 완료 후 엔터")

# iframe 진입
iframe = driver.find_element("tag name", "iframe")

driver.switch_to.frame(iframe)

time.sleep(3)

html = driver.page_source

soup = BeautifulSoup(html, "html.parser")

print(soup.prettify()[:5000])

title = soup.find("meta", {"property": "og:title"})
description = soup.find("meta", {"property": "og:description"})

print(title)
print(description)

input("종료")
driver.quit()
```

```python
title_tag = soup.find("meta", {"property": "og:title"})

description_tag = soup.find("meta", {"property": "og:description"})

title = title_tag["content"]
description = description_tag["content"]

print("제목:")
print(title)

print("\n본문:")
print(description)
```
